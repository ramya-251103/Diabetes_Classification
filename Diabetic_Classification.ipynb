```python
# Distance formula between two numpy arrays - Euclidean distance
def distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn(x, y, querypoint, k=5):
    # Pick K Nearest Neighbours
    vals = []
    
    # For every point in x
    for i in range(x.shape[0]):
        # Compute distance
        d = distance(querypoint, x[i])  # Distance between querypoint and x[i]
        vals.append((d, y[i]))  # Append distance and corresponding label

    # Sort the array and find k nearest points
    vals = sorted(vals, key=lambda x: x[0])  # Sort based on distance (first element of tuple)
    vals = vals[:k]  # First k points
    
    # Majority vote
    vals = np.array(vals, dtype=object)  # Use dtype=object to avoid shape issues
    labels = vals[:, 1]  # Get the labels
    new_values, counts = np.unique(labels, return_counts=True)  # Get unique labels and their counts

    index = counts.argmax()

    pred = new_values[0][index]
    return pred
```

```python
pred = knn(x_train,y_train,x_test)
```
